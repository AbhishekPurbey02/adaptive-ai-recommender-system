```python
import pandas as pd
```

```python
dataset = pd.read_csv('../data/amazon_products.csv',nrows=50000)
```

```python
popular = dataset.sort_values('boughtInLastMonth', ascending= False)
top_10 = popular[['title','stars','price','boughtInLastMonth']].head(10)
print("Top 10 Popular Products: ")
print(top_10)
```

```python
def recommend_popular(n=10):
    return dataset.nlargest(n, 'boughtInLastMonth')[['title','stars','price']]

print(recommend_popular(10))
```
