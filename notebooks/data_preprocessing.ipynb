```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

dataset = pd.read_csv('../data/amazon_products.csv',nrows=5000)
print("Data loaded!")
print(f"Rows: {len(dataset)}")
print(f"Columns: {dataset.columns.to_list()}")
print(dataset.head())
```

```python
small_df = dataset.head(5000)
small_df.to_csv('../data/amazon_small.csv', index=False)
```

```python
# Basic dataset info
print("=== Dataset INFO ===\n")
print(f"Rows: {len(dataset)}")
print(f"Columns: {len(dataset.columns)}")
print(f"\n Missing values: \n {dataset.isnull().sum()}")
```

```python
# Rating Distrubution (Stars)
print("=== Rating Distribution ===\n")
print(dataset['stars'].value_counts())
```

```python
# Popular Producst
print("=== Top 10  Most Popular  Products ===\n")
popular = dataset.nlargest(10, 'boughtInLastMonth')[['title', 'boughtInLastMonth', 'stars', 'price']]
print(popular)
```

```python
# Category breakdown
print("=== Proucts by category ===\n")
print(dataset['category_id'].value_counts().head(10))
```

```python
# Best sellers
print("=== BestSellers ===\n")
bestsellers = dataset[dataset['isBestSeller'] == True]
print(f"Number of bestsellers: {len(bestsellers)}")
print(f"Percentage: {len(bestsellers)/len(dataset)*100:.1f}%")
```
